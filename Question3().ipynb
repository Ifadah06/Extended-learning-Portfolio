```python
from newspaper import Article
import newspaper

def extract_article_info(url):
    try:
        # Create a Article object and download the Article's content
        article = Article(url)
        article.download()
        
        # Parse the downloaded content to extract the Article's title and text
        article.parse()
        
        # Create a dictionary to store the article information
        article_info = {
            'url': url,
            'title': article.title,
            'text': article.text
        }
        
        return article_info
    except Exception as e:
        # Print an error message if an extraction error occurs,
        # then return None to indicate extraction failure.
        print(f"Error extracting article from URL: {url}")
        print(e)
        return None

def extract_articles(urls):
    articles = []
    
    for url in urls:
        # For each URL, use the extract_article_info function
        # and append the article information to the articles list
        article_info = extract_article_info(url)
        
        if article_info:
            articles.append(article_info)
    
    return articles

# Example usage:
urls = [
    'https://www.thetimes.co.uk/article/two-children-stabbed-death-home-stoke-on-trent-lb7ntdlkd',
    'https://www.bbc.com/news/live/world-europe-65872310',
    'https://news.mongabay.com/2023/06/learning-to-live-with-and-love-bears-and-eagles-in-colombias-cloud-forest/',
]

articles = extract_articles(urls)

for article in articles:
    # Print each extracted article's URL, title, and text.
    print(f"URL: {article['url']}")
    print(f"Title: {article['title']}")
    print(f"Text: {article['text']}")
    print()
```
